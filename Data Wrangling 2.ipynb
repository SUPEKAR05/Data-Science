```python
Q1=np.quantile(df['math score'],0.25)
Q3=np.quantile(df['math score'],0.75)
IQR=Q3-Q1
print("Q1 :: ",Q1)
print("Q3 :: ",Q3)
print("IQR :: ",IQR)
upper_2 = Q3 + 1.5*IQR
lower_2 = Q1 - 1.5*IQR
print("lower bound :: ",lower_2)
print("upper bound :: ",upper_2)
```

```python
outliers = (df['math score'] < lower_2) | (df['math score'] > upper_2)

```

```python
print(df[outliers])
```
