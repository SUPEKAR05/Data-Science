```python
# Convert 'Sem1' to numeric with NaN handling
df['Sem1'] = pd.to_numeric(df['Sem1'], errors='coerce')
```

```python
# Fill NaN values with a specific value (e.g., 0)
df['Sem1'].fillna(0, inplace=True)
```

```python
# Convert 'Sem1' to integer
df['Sem1'] = df['Sem1'].astype(int)
```

```python
label_encoder = LabelEncoder()
df['gender_encoded'] = label_encoder.fit_transform(df['Gender'])
```

```python
df['Gender'] = df['gender_encoded']
```
